### Univariate Analysis

```python
def checkPer(dataset, col):
    # Group by the specified column and calculate counts
    counts = dataset.groupby(col).size().reset_index(name='count')

    # Calculate percentages by dividing counts by the total number of rows
    total_rows = len(dataset)
    counts['percentage'] = counts['count'] / total_rows

    return counts
```

```python

result = checkPer(df, 'Occupancy')

# Display the result
print(result)
```

## Converting CreditScore into credit score ranges

***The ranges are in Descending order***

800 to 850: Excellent
Individuals in this range are considered to be low-risk borrowers. They may have an easier time securing a loan than borrowers with lower scores.

740 to 799: Very good
Individuals in this range have demonstrated a history of positive credit behavior and may have an easier time being approved for additional credit.

670 to 739: Good
Lenders generally view those with credit scores of 670 and up as acceptable or lower-risk borrowers.

580 to 669: Fair
Individuals in this category are often considered “subprime” borrowers. Lenders may consider them higher-risk, and they may have trouble qualifying for new credit.

300 to 579: Poor
Individuals in this range often have difficulty being approved for new credit. If you find yourself in the poor category, it's likely you'll need to take steps to improve your credit scores before you can secure any new credit.

```python
# Define the credit score ranges and corresponding labels
credit_score_ranges = [300, 580, 670, 740, 800, 850]
credit_score_labels = ['Poor (300-579)', 'Fair (580-669)', 'Good (670-739)', 'Very Good (740-799)', 'Excellent (800-850)']

# Create the 'CreditScoreRange' column with credit score ranges
df['CreditScoreRange'] = pd.cut(df['CreditScore'], bins=credit_score_ranges, labels=credit_score_labels, right=False)

# Calculate the frequency of each credit score range
credit_score_freq = df['CreditScoreRange'].value_counts()

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(credit_score_freq.index, credit_score_freq.values, color='skyblue')

# Add labels and title
plt.xlabel('Frequency')
plt.ylabel('Credit Score Ranges')
plt.title('Credit Score Ranges Frequency')

# Display the plot
plt.tight_layout()
plt.show()
```

## Converting LTV into LTV_Category

```python
# Define the bins and labels for the categories
bins = [0, 25, 50, 1000]
labels = ['Low', 'Medium', 'High']

# Create a new column 'LTV_Category' with the transformed categories
df['LTV_Category'] = pd.cut(df['LTV'], bins=bins, labels=labels)

# Display the DataFrame with the new 'LTV_Category'
print(df[['LTV', 'LTV_Category']])
```

```python
# Count the occurrences of each category in 'LTV_Category'
category_counts = df['LTV_Category'].value_counts()

# Create a bar plot
plt.figure(figsize=(6, 4))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='pastel')
plt.xlabel('LTV Category')
plt.ylabel('Count')
plt.title('Distribution of LTV Categories')
plt.show()
```

## Converting MonthsInRepayment into Years Range

```python
# Create a new column 'YearsInRepayment' by dividing 'MonthsInRepayment' by 12
df['YearsInRepayment'] = df['MonthsInRepayment'] / 12

# Define the ranges for years
bins = [0, 4, 8, 12, 16, 20] 
# Create labels for each range
labels = ['0-4 years', '5-8 years', '9-12 years', '12-16 years', '16-20 years']

# Cut the 'YearsInRepayment' into the defined ranges and assign the labels
df['YearsRange'] = pd.cut(df['YearsInRepayment'], bins=bins, labels=labels)

# Drop the 'YearsInRepayment' column if not needed anymore
df.drop(columns=['YearsInRepayment'], inplace=True)

# Display the transformed DataFrame
print(df)
```

```python
# Create a bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x='YearsRange', data=df, palette='viridis')
plt.xlabel('Years Range')
plt.ylabel('Count')
plt.title('Bar Plot: Distribution of Months in Repayment (Year Ranges)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
